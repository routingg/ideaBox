"2025-01-15"  join()  

```python
s = "   2025/01/15   "

a = s.strip().split("/")
year = [a[0]]
month = [a[1].zfill(2)]
day = [a[2].zfill(2)]
date = year+month+day

"-".join(date)
```

```python
s = "   2025/01/15   "
y, m, d = s.strip().split("/")
date = "-".join([y, m.zfill(2), d.zfill(2)])
print(date)
```

```python
print("-".join([
    part.zfill(2) if i > 0 else part
    for i, part in enumerate("   2025/01/15   ".strip().split("/"))
]))
```
